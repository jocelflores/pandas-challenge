```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data['SN'].nunique()
df_players = pd.DataFrame({"Total Players":[total_players]}, index=[1])
df_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data['Item Name'].nunique()
avg_price = purchase_data['Price'].round(2).mean()
num_purchases = purchase_data['Purchase ID'].nunique()
total_revenue = purchase_data['Price'].sum()

df_summary = pd.DataFrame({"Number of Unique Items": [unique_items], 
                           "Average Price": [avg_price], 
                          "Number of Purchases": [num_purchases],
                           "Total Revenue": [total_revenue]},index=[1])
df_summary.round({"Average Price":2})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# This does not add up to the total number of players
# gender_count = purchase_data['Gender'].value_counts()
# gender_df = pd.DataFrame(gender_count)
gender_count = purchase_data[['Gender', 'SN']].drop_duplicates()
gender_df = gender_count['Gender'].value_counts()
gender_list = [gender_df[0], gender_df[1], gender_df[2]]
gender_percent = [100*gender_df[0]/total_players,
                  100*gender_df[1]/total_players,
                  100*gender_df[2]/total_players]
gender_table = pd.DataFrame({'Total Count':gender_list,
                            'Percentage of Players': gender_percent},
                           index=['Male', 'Female', 'Other / Non-Disclosed'])
gender_table.round(2)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_analysis = purchase_data.groupby(['Gender'])
count_by_gender = gender_analysis['SN'].count()
avg_price_gender = gender_analysis['Price'].mean()
gender_purchase = gender_analysis['Price'].sum()


#check values individually
#count_by_gender
#avg_price_gender
#gender_purchase

summary_gender = pd.DataFrame({'Purchase Count': count_by_gender,
                        'Average Purchase Price': avg_price_gender,
                        'Total Purchase Value': gender_purchase},
                        index=['Female', 'Male','Other / Non-Disclosed'])

summary_gender
```
